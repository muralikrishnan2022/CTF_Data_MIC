```python
import pandas as pd

# Assuming you have a DataFrame named 'df'
df= pd.read_csv('/content/FIFA 19 data.csv')

# Example using a DataFrame
num_rows, num_columns = df.shape

print(f"Number of Rows: {num_rows}")
print(f"Number of Columns: {num_columns}")
```

```python
df
```

```python
rated_80_players = df[df['Overall'] == 80]
total_rated_80_players = len(rated_80_players)
print(f'Total number of 80-rated players in FIFA 19: {total_rated_80_players}')
```

```python
# Filter Spanish players (assuming 'Nationality' is the column with nationality information)
spanish_players = df[df['Nationality'] == 'Spain']

# Get the total number of Spanish players
total_spanish_players = len(spanish_players)

# Print the result
print(f'Total number of Spanish players in FIFA 19: {total_spanish_players}')
```

```python
filtered_players = df[(df['Value'] >= 10000) & (df['Value'] <= 20000)]

# Check if there are players in the specified range
if not filtered_players.empty:
    # Calculate the mean value of players within the range
    mean_value = round(filtered_players['Value'].mean(), 2)
    print(f'Average value of players between 10,000 and 20,000 coins: {mean_value}')
else:
    print('No players found in the specified value range.')
```

```python
import pandas as pd

# Load your Hogwarts Houses dataset
# Replace 'hogwarts_dataset.csv' with the actual filename or path to your dataset
hogwarts_data = pd.read_csv('/content/harry_potter.csv')

# Assuming your dataset has columns like 'House' and 'Kindness'
# Calculate the mean kindness for each house
average_kindness_per_house = hogwarts_data.groupby('House')['Kindness'].mean()

# Find the house with the highest average kindness
house_with_highest_kindness = average_kindness_per_house.idxmax()
highest_kindness_value = average_kindness_per_house.max()

# Display the result
rounded_off_mean_kindness = round(highest_kindness_value, 2)
print(f"{house_with_highest_kindness}, {rounded_off_mean_kindness}")
```

```python

slytherin_data = hogwarts_data[hogwarts_data['House'] == 'Slytherin']

# Check if there is any data for Slytherin
if not slytherin_data.empty:
    # Calculate the mean affinity to Dark Arts for Slytherin
    average_affinity_slytherin = slytherin_data['Affinity to Dark Arts'].mean()

    # Display the result
    rounded_off_affinity_slytherin = round(average_affinity_slytherin, 2)
    print(f"The average affinity to Dark Arts for Slytherin house is: {rounded_off_affinity_slytherin}")
else:
    print("No data available for Slytherin house.")
```
